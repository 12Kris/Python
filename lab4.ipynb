```python
def get_event(year):
    event = time_dict.get(year)
    if event:
        return event
    else:
        return "Подія не знайдена"

def eventAdd(year, event):
    time_dict[year] = event
    print(f"Подію на {year} було додано.")

def eventRemove(year):
    if year in time_dict:
        del time_dict[year]
        print(f"Подію на {year} було видалено.")

def travel():
    for year in years_to_visit:
        event = get_event(year)
        print(f"У {year} відбулася подія: {event}")

myevents = [(2020, "Пожежі в Австралії"),
            (2021, "Святкування 30-ї Незалежності України"),
            (2022, "Смерть королеви та 'хаос' із прем'єрами"),
            (2023, "Благодійний ярмарок Сиру і Вина у Львові - Lviv Cheese and Wine Festival")]

time_dict = {year: event for year, event in myevents}

years_to_visit = {2020, 2021, 2022, 2023, 2024}

while True:
    print("\nМеню:")
    print("1. Додати подію")
    print("2. Видалити подію")
    print("3. Подорож у часі")
    print("4. Вийти")
    choice = input("Виберіть опцію (1/2/3/4): ")

    if choice == "1":
        year = int(input("Введіть рік події: "))
        event = input("Введіть опис події: ")
        eventAdd(year, event)
    elif choice == "2":
        year = int(input("Введіть рік події, яку ви хочете видалити: "))
        eventRemove(year)
    elif choice == "3":
        travel()
    elif choice == "4":
        print("Дякуємо за використання програми")
        break
    else:
        print("Неправильний вибір. Спробуйте ще раз.")

```
